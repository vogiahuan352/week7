```python
import pandas as pd 
url='50_Startups.csv'
data=pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
from pandas.core.frame import collections
```

```python
correlation=data.corr()
print(correlation)
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Profit']),y=data['R&D Spend'])
print('He so intercept la',model.intercept_)
print('he so coeffient la',model.coef_)
model.score(X=pd.DataFrame(data['Profit']),y=data['R&D Spend'])
```

```python
y_pred=model.predict(X=pd.DataFrame(data['Profit']))
data.plot(kind='scatter',x='Profit',y='R&D Spend',figsize=(9,9),color='black',)
plt.plot(data['Profit'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X=pd.DataFrame([data['Profit'],data['Profit']**2]).T
model.fit(X,y=data['R&D Spend'])
print('Intercept cua mo hinh:',model.intercept_)
print('Coeffient cua mo hinh:',model.coef_)
print('do chinh xac cua mo hinh:',model.score(X,y=data['R&D Spend']))
```

```python
import numpy as np
```

```python
range=np.arange(1,200000,1)
X_range=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=X_range)
data.plot(kind='scatter',x='R&D Spend', y='Profit',figsize=(9,9),color='black')
plt.plot(range,y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
model.fit(X=data.loc[:,['Profit','R&D Spend']],y=data['Profit'])
print('intercept la:',model.intercept_)
print('coeffient la:',model.coef_)
print('do chinh xac:',model.score(X=data.loc[:,['Profit','R&D Spend']],y=data['Profit']))
```
